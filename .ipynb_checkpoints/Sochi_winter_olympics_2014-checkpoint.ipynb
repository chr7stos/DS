There is a point system in which each country is awarded:
- 4 points for eachgold medal
- 2 points for each silver medal
- 1 point for each bronze medal.  

Olympic_points_df is consisted of:
1.  a column called 'country_name' with the country name
2.  a column called 'points' with the total number of points the countryearned at the Sochi olympics.

```python
point_system = [4, 2, 1]
df["points"] = df[["gold", "silver", "bronze"]].dot(point_system)
olympic_points_df = df[["country_name", "points"]]
olympic_points_df
```

```python
## Using numpy

points = np.dot(df[["gold", "silver", "bronze"]], point_system)
olympic_points = {"country_name": pd.Series(countries), "points": pd.Series(points)}
olympic_points_df = pd.DataFrame(olympic_points)
olympic_points_df
```
